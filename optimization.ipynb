```python
df=pd.read_csv('fashion_products-1.csv')
df.head()
```

```python
df['Color']=df['Color'].str.lower()
df['Brand']=df['Brand'].str.lower()
df['Category']=df['Category'].str.lower()
df['Size']=df['Size'].str.lower()
df['Product Name']=df['Product Name'].str.lower()
```

```python
df['Description']=df1['fulldescription']
```

```python
product_name = pd.Series(df['Product Name'].unique()).str.lower()
brand = pd.Series(df['Brand'].unique()).str.lower()
category = pd.Series(df['Category'].unique()).str.lower()
color = pd.Series(df['Color'].unique()).str.lower()
size = pd.Series(df['Size'].unique()).str.lower()
```

```python
brand
```

```python
nlp=spacy.load('en_core_web_sm')
```

```python
size
```

```python
user_query="running shoes"
query_dict={"Product Name":[],
            "Brand":[],
            "Category":[],
            "Color":[],
            "Size":[]}
other_tokens = []


doc=nlp(user_query)
for ent in doc:
     unique=[]
     if ent.is_stop:
        continue
     print(ent)
     ent= ent.text.lower()
     if ent in product_name.values:
         query_dict['Product Name'].append(ent)
     elif ent in brand.values:
         query_dict['Brand'].append(ent)
     elif ent in category.values:
         query_dict['Category'].append(ent)
     elif ent in color.values:         
         query_dict['Color'].append(ent)
     elif ent in size.values:         
         query_dict['Size'].append(ent)
     else:
        other_tokens.append(ent)
    
```

```python
query_dict = {k: v for k, v in query_dict.items() if v}
query_dict
```

```python
other_tokens
```

```python
if "Color" in query_dict:
        color_filter = df[df['Color'].isin(query_dict['Color'])]
else:
        color_filter=df.copy()
if "Brand" in query_dict:
        #print("brand")
        #print(query_dict['Brand'])
        brand_filter=color_filter[color_filter['Brand'].isin(query_dict['Brand'])]
        #print(brand_filter)
else:
        brand_filter=color_filter
if "Product Name" in query_dict:
        #print('Productfiler')
        product_filter=brand_filter[brand_filter['Product Name'].isin(query_dict['Product Name'])]
        #print(product_filter)
else:
        print('else')
        product_filter=brand_filter

if "Size" in query_dict:
        print('size')
        size_filter=brand_filter[brand_filter['Size'].isin(query_dict['Size'])]

else:
        print('else size')
        size_filter=product_filter
        print(size_filter)


if "Category" in query_dict:
        category_filter=size_filter[size_filter['Cateory'].isin(query_dict['Category'])]

else:
        category_filter=size_filter
finaldf=category_filter
```

```python
category_filter_index=category_filter.index
category_filter_index
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
tokens = ['mens', 'soft', 'fur']
combined = " ".join(tokens)

print(combined)
```

```python
finaldf.reset_index().drop('index',inplace=True,axis=1)
```

```python
if combined:
    descriptions = df['Description'].tolist()
    descriptions.append(other_tokens) 
    tfidf_vectorizer = TfidfVectorizer(stop_words='english')  
    tfidf_matrix = tfidf_vectorizer.fit_transform(df['Description'])

    other_tokens_vector = tfidf_vectorizer.transform([combined])

    # Only compute cosine similarity for rows in the filtered set
    #Slice the tfidf_matrix to only include rows for the filtered categories
    filtered_tfidf_matrix = tfidf_matrix[category_filter_index]
    #print(filtered_tfidf_matrix)

    # Get the cosine similarity between the `other_tokens_text` and all the descriptions
    cosine_similarities = cosine_similarity(other_tokens_vector, filtered_tfidf_matrix).flatten()


    top_10 =np.argsort(cosine_similarities)[::-1][:10]
    print(top_10)


    most_similar_description = finaldf.iloc[top_10]
    
```
